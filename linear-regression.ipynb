```python
import pandas as pd
```

```python
# We will predict sepal width from sepal length using Linear Regression

dataset_path = "iris.csv"

df = pd.read_csv(dataset_path, header=None, names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species'])

X = df.iloc[:, 0].values.astype(float) # sepal length
Y = df.iloc[:, 2].values.astype(float) # petal length
```

```python
# y = mx + c

# m = n * sum(xi * yi) - sum(x) * sum(y) / n * sum(xi^2) - (sum(x))^2
# c = mean(y) - mean(x) * m

n = len(X)
sum_xi_yi = sum(X * Y)
sum_x = sum(X)
sum_y = sum(Y)

sum_xi_sq = sum(X**2)
sum_x_sq = sum_x ** 2

y_mean = sum_y / n
x_mean = sum_x / n

numerator = ( n * sum_xi_yi ) - (sum_x * sum_y)
denominator = ( n * sum_xi_sq ) - sum_x_sq

m = numerator / denominator

c = y_mean - x_mean * m
```

```python
def predict(x):
    global m, c
    return m * x + c
```

```python
predicted_y = list()

for x in X:
    if x != None:
        predicted_y.append(predict(x))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X, Y, color="red", label="Data")
plt.plot(X, predicted_y, color="green", label="Prediction / Regression Line")
plt.xlabel("Input Feature - Sepal Length")
plt.ylabel("Output Feature - Petal Length")
plt.legend()
plt.show()
```

```python
test_x = float(input("Enter input feature - Sepal Length: "))
pred_y = predict(test_x)

print(f"The predicted output feature - Petal Length is {pred_y}")
```

```python
# Errors

# MAE
# sum of absolute ( actual y - predicted y ) / n
mae = sum(abs(Y-predicted_y)) / n

# MSE
# sum of square of (actual y - predicted y) / n
mse = sum((Y-predicted_y)**2) / n

print(f"Mean Absolute Error : {mae} \nMean Sqaure Error : {mse}")
```
